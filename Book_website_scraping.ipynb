Create a For Loop to loop through all results from all pages

```python
title =[]
price=[]
avail = []
for i in range(1,51):
  website = 'https://books.toscrape.com/catalogue/page-'+ str(i) +'.html'
  response = requests.get(website)
  soup = bs(response.content,'html.parser')
  results = soup.find_all('li',{'class':'col-xs-6'})
  for result in results: 
  #title
    try:
      title.append(result.find('h3').find('a').get('title'))
    except:
      title.append('n/a')
  #Price
    try: 
      avail.append(result.find('p',{'class':'price_color'}).get_text())
    except:
      avail.append('n/a')
 
  #Availability
    try:
      price.append(result.find('p',{'class':'instock availability'}).get_text().strip())
    except:
      price.append('n/a')
```

Create a Dataframe

```python
df = pd.DataFrame({'title':title,'price':price,'Availability':avail})
```
