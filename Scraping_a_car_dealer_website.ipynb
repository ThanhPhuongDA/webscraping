```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

HTTP request

```python
website = 'https://www.cars.com/shopping/results/?page=1&page_size=20&list_price_max=&makes[]=mercedes_benz&maximum_distance=20&models[]=&stock_type=cpo&zip='
```

Get Requests

```python
response = requests.get(website)
```

```python
response.status_code
```

Convert HTML into beautifulsoup object

```python
soup = bs(response.content,'html.parser')
```

```python
soup
```

Results

```python
result = soup.find_all('div',{'class':'vehicle-card'})
```

Put everything together inside a For Loop

Name

Mileage

Rating

Price

Car Dealer

```python
name=[]
mileage=[]
rating = []
rating_count = []
car_dealer =[]
price=[]

for info in result:
  #name
  try:
    name.append(info.find('h2',{'class':'title'}).get_text())
  except:
    name.append('n/a')
  #mileage
  try:
    mileage.append(info.find('div',{'class':'mileage'}).get_text())
  except:
    mileage.append('n/a')
  #rating
  try:
    rating.append(info.find('span',{'class':'sds-rating__count'}).get_text())
  except:
    rating.append('n/a')
  # rating_count
  try:
    rating_count.append(info.find('span',{'sds-rating__link sds-button-link'}).get_text())
  except:
    rating_count.append('n/a')
  #Price
  try:
    price.append(info.find('span',{'class':'primary-price'}).get_text())
  except:
    price.append('n/a')
  #Car dealer
  try:
    car_dealer.append(info.find('div',{'class':'dealer-name'}).get_text().strip())
  except:
    car_dealer.append('n/a')
```

```python
rating_count
```

```python
len(rating_count)
```

Create Data Frame

```python
df = pd.DataFrame({'Name':name,'Mileage':mileage,'Dealer_name':car_dealer,'Rating':rating,'Rating_count':rating_count,'Price':price})
```

```python
df
```

Data cleaning

```python
df['Rating_count'] = df['Rating_count'].apply(lambda x:x.strip('reviews)').strip('('))
```

```python
df['Rating_count']
```

```python
df
```

```python
df.to_excel('single_page_car.xlsx',index=False)
```

Pagination to get results from multiple pages

```python
 name=[]
mileage=[]
rating = []
rating_count = []
car_dealer =[]
price=[]
for i in range(1,20):
  website = 'https://www.cars.com/shopping/results/?page='+str(i)+'&page_size=20&list_price_max=&makes[]=mercedes_benz&maximum_distance=20&models[]=&stock_type=cpo&zip='
  response = requests.get(website)
  soup = bs(response.content,'html.parser')
  result = soup.find_all('div',{'class':'vehicle-card'})
  for info in result:
  #name
    try:
      name.append(info.find('h2',{'class':'title'}).get_text())
    except:
      name.append('n/a')
  #mileage
    try:
      mileage.append(info.find('div',{'class':'mileage'}).get_text())
    except:
      mileage.append('n/a')
  #rating
    try:
      rating.append(info.find('span',{'class':'sds-rating__count'}).get_text())
    except:
      rating.append('n/a')
  # rating_count
    try:
      rating_count.append(info.find('span',{'sds-rating__link sds-button-link'}).get_text())
    except:
      rating_count.append('n/a')
  #Price
    try:
      price.append(info.find('span',{'class':'primary-price'}).get_text())
    except:
      price.append('n/a')
  #Car dealer
    try:
      car_dealer.append(info.find('div',{'class':'dealer-name'}).get_text().strip())
    except:
      car_dealer.append('n/a')
```

```python
df1 = pd.DataFrame({'Name':name,'Mileage':mileage,'Dealer_name':car_dealer,'Rating':rating,'Rating_count':rating_count,'Price':price})
```

```python
df1
```

```python
df1['Rating_count'] = df1['Rating_count'].apply(lambda x:x.strip('reviews)').strip('('))
```
